```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

```python
template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

topic = input("Enter topic: ")

print(template_1.format(topic=topic))
```

```python
def multi_input_prompt(topic, audience, tone):

    template_2 = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} tone"
    )

    return template_2.format(topic=topic, audience=audience, tone=tone)
```

```python
topic = input("Enter topic: ")
audience = input("Enter audience: ")
tone = input("Enter tone: ")
```

```python
print(multi_input_prompt(topic, audience, tone))
```

```python
def prompt_variations(topic):

    teaching_template = PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    )

    interview_template = PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 questions about {topic}"
    )

    storytelling_template = PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story"
    )

    print("Teaching Style:")
    print(teaching_template.format(topic=topic))

    print("\nInterview Style:")
    print(interview_template.format(topic=topic))

    print("\nStorytelling Style:")
    print(storytelling_template.format(topic=topic))
```

```python
topic = input("Enter topic: ")
prompt_variations(topic)
```

```python
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher."

    elif role == "interviewer":
        system_message = "You are an interviewer who asks questions."

    elif role == "motivator":
        system_message = "You are a motivational speaker."

    else:
        system_message = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)
```

```python
role = input("Enter role (teacher/interviewer/motivator): ")
topic = input("Enter topic: ")
```

```python
messages = get_chat_prompt(role, topic)

for msg in messages:
    print(msg.content)
```

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default to 'formal'")
        tone = "formal"

    return audience, tone
```

```python
audience = input("Enter audience: ")
tone = input("Enter tone: ")
```

```python
audience, tone = validate_inputs(audience, tone)

print("Final Audience:", audience)
print("Final Tone:", tone)
```

```python
def generate_prompt(topic, audience, tone, style):

    # Validate inputs first
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )
        return template.format(topic=topic, audience=audience, tone=tone)

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 interview questions about {topic}"
        )
        return template.format(topic=topic)

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
        return template.format(topic=topic, audience=audience, tone=tone)

    else:
        return "Invalid style selected"
```

```python
topic = input("Enter topic: ")
audience = input("Enter audience: ")
tone = input("Enter tone: ")
style = input("Enter style (teaching/interview/storytelling): ")
```

```python
result = generate_prompt(topic, audience, tone, style)
print(result)
```
